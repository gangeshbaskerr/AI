3.  function isValid(mapping, problem):
    evaluate the problem using the current mapping
    return True if the problem is satisfied, otherwise False

4.  function extractUniqueLetters(problem):
    extract and return the unique letters from the problem

5.  function printSolution(mapping):
    print the solution

```python
import itertools

def extract_unique_letters(problem):
    unique_letters = set()
    for word in problem:
        for letter in word:
            unique_letters.add(letter)
    return list(unique_letters)

def word_to_number(word, mapping):
    return int(''.join(str(mapping[letter]) for letter in word))

def is_valid(mapping, problem):
    first_word, second_word, result_word = problem[0], problem[1], problem[2]
    
    first_number = word_to_number(first_word, mapping)
    second_number = word_to_number(second_word, mapping)
    result_number = word_to_number(result_word, mapping)

    return first_number + second_number == result_number

def solve(letters, used_digits, mapping, index, problem):
    if index == len(letters):
        return is_valid(mapping, problem)
    
    for digit in range(10):
        if digit == 0 and letters[index] in [word[0] for word in problem]:
            continue
        
        if not used_digits[digit]:
            used_digits[digit] = True
            mapping[letters[index]] = digit
            if solve(letters, used_digits, mapping, index + 1, problem):
                return True
            used_digits[digit] = False
            mapping[letters[index]] = None
    
    return False

def solve_cryptarithmetic(problem):
    letters = extract_unique_letters(problem)
    if len(letters) > 10:
        print("Too many letters in the problem.")
        return 
    used_digits = [False] * 10
    mapping = {letter: None for letter in letters}
    if solve(letters, used_digits, mapping, 0, problem):
        print_solution(mapping)
    else:
        print("No solution found.")

def print_solution(mapping):
    print("Solution found:")
    for letter, digit in mapping.items():
        print(f"{letter} = {digit}")

problem = [
    "TWO",  # First operand
    "TWO",  # Second operand
    "FOUR"  # Result
]

solve_cryptarithmetic(problem)
```
